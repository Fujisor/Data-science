```python
import pandas as pd
```

```python
# create a data

import random

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Mark", "Rannie"],
    "Math": [85, 60, 90, 70, 88, 92, 85, 96],
    "Science": [random.randint(75, 100) for _ in range(8)], # Will generate a random 8 values
    "English": [random.randint(75, 100) for _ in range(8)]
}
```

```python
df_student_grade = pd.DataFrame(data) # will create a dataframe
```

```python
# show data frame 
df_student_grade
```

```python
# select all rows where math > 80 (using .loc)
df_student_grade.loc[df_student_grade['Math'] > 80, :]
```

```python
# Select the first 3 students using .iloc
df_student_grade.iloc[0:4]
```

```python
# Use .isin() to find/filter students who scored either 75 or 90 in English
df_student_grade[df_student_grade['English'].isin([75, 90])]
```

```python
# Add a new column Average = row-wise average of subjects.
df_student_grade['Average'] = df_student_grade.apply(lambda column_grade: 
                                                     (column_grade['Math'] + column_grade['Science'] +column_grade['English'])/3, axis=1)
                                                      # axis = 1 to give access to column
```

```python
# Use .apply() with a lambda to classify each student as "Pass" if Average >= 75, else "Fail".
df_student_grade['Result'] = df_student_grade.apply(lambda column_average: "Pass" if column_average['Average'] >= 75 else "Fail" , axis=1)
```

```python
# show dataframe 
df_student_grade[['Name', 'Average', 'Result']]
```
